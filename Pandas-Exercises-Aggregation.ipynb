# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
data = pd.read_csv('occupation.csv',sep='|')
```

### Step 3. Assign it to a variable called users.

```python
users=data
users
```

### Step 4. Discover what is the mean age per occupation

```python
b1 = users.groupby(by=['occupation'])[['age']].mean().add_prefix('Mean_of_').reset_index()
b1
```

```python
b1 = users.groupby(by=['occupation'])[['age']].mean().add_prefix('Mean_of_').reset_index()
b1
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
x=users.groupby(['occupation','gender'])['gender'].count()
y=users.groupby(['occupation'])['gender'].count()
r=((x/y)*100).round(2)
print(r)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
max_age= users.groupby('occupation').aggregate({'age': 'max'})
min_age=users.groupby('occupation').aggregate({'age': 'min'})
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age= users.groupby(by=['occupation','gender'])[['age']].mean().add_prefix('Mean_of_').reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
new=users.groupby(['occupation','gender']).agg({'gender': 'count'})
new.apply(lambda x: x/x.sum())
```
